```python
Train_df = Train_df.drop(['Price'],axis=1)
Train_df
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(Train_df,target)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV,cross_val_score
clf = {
    
    "extrtreeg":{
        "model": ExtraTreesRegressor(),
        "params":{}
    },
    
    "RandmForest":{
        "model":RandomForestRegressor(),
        "params":{}
    },
    
    "LinReg":{
        "model":LinearRegression(),
        "params":{}
    }
}
```

```python
scores = []
for model_name, modell in clf.items():
    gsc = GridSearchCV(modell['model'],modell['params'],cv=5)
    gsc.fit(X_train,y_train)
    scoreb= gsc.score(X_test,y_test)
    cvs = cross_val_score(modell['model'],X_train,y_train)
    scores.append({
        "model_nam":model_name,
        "cross_val_sc": cvs,
        "score":scoreb
    })
```
